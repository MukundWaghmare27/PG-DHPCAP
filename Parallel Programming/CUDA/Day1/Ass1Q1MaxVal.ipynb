```python
%%writefile maxVal.cu

#include <stdio.h>
#include <cuda_runtime.h>
#include <device_launch_parameters.h>

__global__ void intnum(int *a, int *b, int *c, int *max_val) {
    if (*a > *b && *a > *c) {
        *max_val = *a;
    }
    else if (*b > *a && *b > *c) {
        *max_val = *b;
    }
    else {
        *max_val = *c;
    }
}

int main() {
    int host_a = 11;
    int host_b = 55;
    int host_c = 33;
    int host_max_val;

    int *device_a, *device_b, *device_c, *device_max_val;

    cudaMalloc((void **)&device_a, sizeof(int));
    cudaMalloc((void **)&device_b, sizeof(int));
    cudaMalloc((void **)&device_c, sizeof(int));
    cudaMalloc((void **)&device_max_val, sizeof(int));

    cudaMemcpy(device_a, &host_a, sizeof(int), cudaMemcpyHostToDevice);
    cudaMemcpy(device_b, &host_b, sizeof(int), cudaMemcpyHostToDevice);
    cudaMemcpy(device_c, &host_c, sizeof(int), cudaMemcpyHostToDevice);

    intnum<<<1, 1>>>(device_a, device_b, device_c, device_max_val);

    cudaMemcpy(&host_max_val, device_max_val, sizeof(int), cudaMemcpyDeviceToHost);
    printf("Max value is %d\n", host_max_val);

    cudaFree(device_a);
    cudaFree(device_b);
    cudaFree(device_c);
    cudaFree(device_max_val);

    return 0;
}
```
